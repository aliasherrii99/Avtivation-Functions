```python
import numpy as np
import matplotlib.pyplot as plt
    
class ActivationFunction:
    def __init__(self, function ,x):
        self.function = function
        self.x = x
        
    def relu(self):
        mask = self.x > 0
        der_value = mask.astype(float)
        fun_value = self.x * der_value
        return fun_value , der_value
        
        
    def pureline(self):
        return self.x , np.sign(self.x)
    
    def sigmoid(self):
        function_value = 1/(1+np.exp(-self.x))
        return function_value , function_value * (1-function_value )
    
    def tanh(self):
        return np.tanh(self.x) , 1 - np.tanh(self.x) ** 2
    
    def activation(self):
        
        if self.function == 'sigmoid':
            return self.sigmoid() 
        elif self.function == 'tanh':
            return self.tanh()
        elif self.function == 'relu':
            return self.relu()
        elif self.function == 'pureline':
            return self.pureline()
        else:
            return 'incorrect input'
        
        
    def draw_plot(self):
        
        
        xpoints = self.x
        ypoints = self.activation()[0]
        
        fig , axs =  plt.subplots(1 , 2 , figsize = (12,4))
        

        axs[0].set_title(self.function)
        axs[0].grid(True)
        axs[0].plot(xpoints,ypoints)
        
        ypoints2 = self.activation()[1]
        axs[1].set_title(f"derivative of {self.function}")
        axs[1].grid(True)
        axs[1].plot(xpoints,ypoints2)

        

xx = np.arange(-10, 10, 0.1)
net1 = ActivationFunction('relu',xx)
net1.draw_plot()

net2 = ActivationFunction('sigmoid',xx)
net2.draw_plot()

net3 = ActivationFunction('tanh',xx)
net3.draw_plot()

net4 = ActivationFunction('pureline',xx)
net4.draw_plot()
# net2 = ActivationFunction()
```
